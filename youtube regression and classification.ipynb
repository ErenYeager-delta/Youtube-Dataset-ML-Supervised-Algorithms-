```python
import pandas as pd
data=pd.read_csv('youtube-top-100-songs-2025.csv')
```

```python
print(data.isnull().sum())
```

```python
print(data.dropna(inplace=True))
```

```python
print(data.duplicated().sum())
```

```python
print(data.drop_duplicates(inplace=True))
```

```python
print(data.info())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols=['title','fulltitle','description','categories','tags','duration_string','thumbnail','channel','channel_url','live_status']
for col in cols:
    data[col]=le.fit_transform(data[col])
    print(data[col])
```

```python
print(data.info())
```

```python
from sklearn.preprocessing import st
```

```python
# data feature and Data target
X=data[['fulltitle']]
y=data['channel_follower_count']

#data spliting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 32)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression

model_1= LinearRegression()
model_1.fit(X_train,y_train)
y_pred_1=model_1.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_absolute_error

# For Model_1 Evalution Matrics

Mse= mean_squared_error(y_test, y_pred_1 )
print (f'Mse for Model 1: {Mse * 100: .2f}%')

r2= r2_score(y_test, y_pred_1)
print(f'r2 for Model 1: {r2* 100: .2f}%')

Rmse= root_mean_squared_error(y_test, y_pred_1)
print(f'Rmse for Model 1: {Rmse * 100: .2f}%')

Mae= mean_absolute_error(y_test, y_pred_1)
print(f'Mae for Model 1: {Mae * 100: .2f}%')
```

```python
from sklearn.svm import SVR
model_2= SVR()
model_2.fit(X_train,y_train)
y_pred_2=model_2.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_absolute_error

# For Model_2 Evalution Matrics

Mse= mean_squared_error(y_test, y_pred_2)
print(f'Mse for Model 2: {Mse * 100: .2f}%')

r2= r2_score(y_test, y_pred_2)
print(f'r2 for Model 2: {r2* 100: .2f}%')

Rmse= root_mean_squared_error(y_test, y_pred_2)
print(f'Rmse for Model 2: {Rmse * 100: .2f}%')

Mae= mean_absolute_error(y_test, y_pred_2)
print(f'Mae for Model 2: {Mae * 100: .2f}%')
```

```python
from sklearn.tree import DecisionTreeRegressor
model_3= DecisionTreeRegressor()
model_3.fit(X_train,y_train)
y_pred_3=model_3.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_absolute_error

# For Model_3 Evalution Matrics

Mse= mean_squared_error(y_test, y_pred_3)
print(f'Mse for Model 3: {Mse * 100: .2f}%')

r2= r2_score(y_test, y_pred_3)
print(f'r2 for Model 3: {r2* 100: .2f}%')

Rmse= root_mean_squared_error(y_test, y_pred_3)
print(f'Rmse for Model 3: {Rmse * 100: .2f}%')

Mae= mean_absolute_error(y_test, y_pred_3)
print(f'Mae for Model 3: {Mae * 100: .2f}%')
```

```python
from sklearn.ensemble import RandomForestRegressor

model_4= RandomForestRegressor()
model_4.fit(X_train,y_train)
y_pred_4=model_4.predict(X_test)

#Evalution Metrics for all algorthms ( Mean square error, R_square, Root Mean square error, Mean Absolute error )

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_absolute_error

# For Model_4 Evalution Matrics

Mse= mean_squared_error(y_test, y_pred_4)
print(f'Mse for Model 4: {Mse * 100: .2f}%')

r2= r2_score(y_test, y_pred_4)
print(f'r2 for Model 4: {r2* 100: .2f}%')

Rmse= root_mean_squared_error(y_test, y_pred_4)
print(f'Rmse for Model 4: {Rmse * 100: .2f}%')

Mae= mean_absolute_error(y_test, y_pred_4)
print(f'Mae for Model 4: {Mae * 100: .2f}%')
```
